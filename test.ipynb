```python
class BobailGame:
    def __init__(self):
        self.board = [[" " for _ in range(5)] for _ in range(5)]  # Plateau 5x5
        self.setup_board()

    def setup_board(self):
        # Placer les pions des joueurs
        for i in range(5):
            self.board[0][i] = "1"  # Pions du joueur 1
            self.board[4][i] = "2"  # Pions du joueur 2
        # Placer le Bobail
        self.board[2][2] = "B"

    def move_up(self, x, y):
        for i in range(x - 1, -1, -1):
            if self.board[i][y] != " ":
                return i + 1, y
        return 0, y

    def move_down(self, x, y):
        for i in range(x + 1, 5):
            if self.board[i][y] != " ":
                return i - 1, y
        return 4, y

    def move_left(self, x, y):
        for j in range(y - 1, -1, -1):
            if self.board[x][j] != " ":
                return x, j + 1
        return x, 0

    def move_right(self, x, y):
        for j in range(y + 1, 5):
            if self.board[x][j] != " ":
                return x, j - 1
        return x, 4

    def move_diagonal_up_left(self, x, y):
        step_x, step_y = -1, -1
        while x + step_x >= 0 and y + step_y >= 0:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x + 1, y + step_y + 1
            x += step_x
            y += step_y
        return max(x + step_x, 0), max(y + step_y, 0)

    def move_diagonal_up_right(self, x, y):
        step_x, step_y = -1, 1
        while x + step_x >= 0 and y + step_y < 5:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x + 1, y + step_y - 1
            x += step_x
            y += step_y
        return max(x + step_x, 0), min(y + step_y, 4)

    def move_diagonal_down_left(self, x, y):
        step_x, step_y = 1, -1
        while x + step_x < 5 and y + step_y >= 0:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x - 1, y + step_y + 1
            x += step_x
            y += step_y
        return min(x + step_x, 4), max(y + step_y, 0)

    def move_diagonal_down_right(self, x, y):
        step_x, step_y = 1, 1
        while x + step_x < 5 and y + step_y < 5:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x - 1, y + step_y - 1
            x += step_x
            y += step_y
        return min(x + step_x, 4), min(y + step_y, 4)

    def select_and_move(self):
        self.display_board()
        start = tuple(map(int, input("Coordonnées de départ (x y) : ").split()))
        direction = input("Direction du déplacement (up, down, left, right, diagonal_up_left, diagonal_up_right, diagonal_down_left, diagonal_down_right) : ").strip().lower()

        x, y = start
        if self.board[x][y] not in ["1", "2", "B"]:
            print("Aucune pièce valide à cet endroit.")
            return

        if direction == "up":
            new_x, new_y = self.move_up(x, y)
        elif direction == "down":
            new_x, new_y = self.move_down(x, y)
        elif direction == "left":
            new_x, new_y = self.move_left(x, y)
        elif direction == "right":
            new_x, new_y = self.move_right(x, y)
        elif direction == "diagonal_up_left":
            new_x, new_y = self.move_diagonal_up_left(x, y)
        elif direction == "diagonal_up_right":
            new_x, new_y = self.move_diagonal_up_right(x, y)
        elif direction == "diagonal_down_left":
            new_x, new_y = self.move_diagonal_down_left(x, y)
        elif direction == "diagonal_down_right":
            new_x, new_y = self.move_diagonal_down_right(x, y)
        else:
            print("Direction invalide.")
            return

        piece = self.board[x][y]  # Conserver la pièce déplacée
        self.board[x][y] = " "  # Vider la case de départ
        self.board[new_x][new_y] = piece  # Déplacer la pièce à la nouvelle position

    def display_board(self):
        for row in self.board:
            print(" | ".join(row))
            print("-" * 19)

# Lancer le jeu
if __name__ == "__main__":
    game = BobailGame()
    while True:
        game.select_and_move()
```

```python
class BobailGame:
    def __init__(self):
        self.board = [[" " for _ in range(5)] for _ in range(5)]  # Plateau 5x5
        self.setup_board()
        self.current_player = "1"  # Le joueur 1 commence

    def setup_board(self):
        # Placer les pions des joueurs
        for i in range(5):
            self.board[0][i] = "1"  # Pions du joueur 1
            self.board[4][i] = "2"  # Pions du joueur 2
        # Placer le Bobail
        self.board[2][2] = "B"

    def display_board(self):
        color_map = {
            "1": "\033[92m1\033[0m",  # Vert pour les pions du joueur 1
            "2": "\033[91m2\033[0m",  # Rouge pour les pions du joueur 2
            "B": "\033[93mB\033[0m",  # Jaune pour le Bobail
            " ": " "
        }
        for row in self.board:
            print(" | ".join(color_map[cell] for cell in row))
            print("-" * 19)

    def move_up(self, x, y):
        for i in range(x - 1, -1, -1):
            if self.board[i][y] != " ":
                return i + 1, y
        return 0, y

    def move_down(self, x, y):
        for i in range(x + 1, 5):
            if self.board[i][y] != " ":
                return i - 1, y
        return 4, y

    def move_left(self, x, y):
        for j in range(y - 1, -1, -1):
            if self.board[x][j] != " ":
                return x, j + 1
        return x, 0

    def move_right(self, x, y):
        for j in range(y + 1, 5):
            if self.board[x][j] != " ":
                return x, j - 1
        return x, 4

    def move_diagonal_up_left(self, x, y):
        step_x, step_y = -1, -1
        while x + step_x >= 0 and y + step_y >= 0:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x + 1, y + step_y + 1
            x += step_x
            y += step_y
        return max(x + step_x, 0), max(y + step_y, 0)

    def move_diagonal_up_right(self, x, y):
        step_x, step_y = -1, 1
        while x + step_x >= 0 and y + step_y < 5:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x + 1, y + step_y - 1
            x += step_x
            y += step_y
        return max(x + step_x, 0), min(y + step_y, 4)

    def move_diagonal_down_left(self, x, y):
        step_x, step_y = 1, -1
        while x + step_x < 5 and y + step_y >= 0:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x - 1, y + step_y + 1
            x += step_x
            y += step_y
        return min(x + step_x, 4), max(y + step_y, 0)

    def move_diagonal_down_right(self, x, y):
        step_x, step_y = 1, 1
        while x + step_x < 5 and y + step_y < 5:
            if self.board[x + step_x][y + step_y] != " ":
                return x + step_x - 1, y + step_y - 1
            x += step_x
            y += step_y
        return min(x + step_x, 4), min(y + step_y, 4)

    def check_victory(self):
        # Vérifier si le Bobail atteint la ligne de départ d'un joueur
        for i in range(5):
            if self.board[0][i] == "B":  # Ligne de départ du joueur 1
                print("Le joueur 1 a gagné !")
                return True
            if self.board[4][i] == "B":  # Ligne de départ du joueur 2
                print("Le joueur 2 a gagné !")
                return True
        return False

    def select_and_move(self):
        print(f"C'est au tour du joueur {self.current_player}.")
        start = tuple(map(int, input("Coordonnées de départ (x y) : ").split()))
        x, y = start

        # Vérifier que le joueur ne peut déplacer que ses pions
        if self.board[x][y] != self.current_player and self.board[x][y] != "B":
            print("Vous ne pouvez déplacer que vos propres pions ou le Bobail.")
            return

        direction = input("Direction du déplacement (up, down, left, right, diagonal_up_left, diagonal_up_right, diagonal_down_left, diagonal_down_right) : ").strip().lower()

        if direction == "up":
            new_x, new_y = self.move_up(x, y)
        elif direction == "down":
            new_x, new_y = self.move_down(x, y)
        elif direction == "left":
            new_x, new_y = self.move_left(x, y)
        elif direction == "right":
            new_x, new_y = self.move_right(x, y)
        elif direction == "diagonal_up_left":
            new_x, new_y = self.move_diagonal_up_left(x, y)
        elif direction == "diagonal_up_right":
            new_x, new_y = self.move_diagonal_up_right(x, y)
        elif direction == "diagonal_down_left":
            new_x, new_y = self.move_diagonal_down_left(x, y)
        elif direction == "diagonal_down_right":
            new_x, new_y = self.move_diagonal_down_right(x, y)
        else:
            print("Direction invalide.")
            return

        piece = self.board[x][y]  # Conserver la pièce déplacée
        self.board[x][y] = " "  # Vider la case de départ
        self.board[new_x][new_y] = piece  # Déplacer la pièce à la nouvelle position

        # Vérifier si quelqu'un a gagné
        if self.check_victory():
            exit()

        # Passer au joueur suivant
        self.current_player = "2" if self.current_player == "1" else "1"

        # Afficher le plateau après le déplacement
        self.display_board()

# Lancer le jeu
if __name__ == "__main__":
    game = BobailGame()
    while True:
        game.select_and_move()
```
